```python
# import necessary stuff

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tsfresh
import os
```

```python
income_table_path = '../census-google_trends-matching/data/metro_areas_table_3.csv'
income_table = pd.read_csv(income_table_path)
income_table.head()
```

```python
google_trends_data_path = '../google-trends-retrieval/data/trends_data_2.json'
google_trends_data = pd.read_json(google_trends_data_path)
google_trends_data.head()
```

```python
categories = [1220, 38, 60]#, 310, 899, 229, 277, 75, 366, 59, 54, 29, 78, 67, 1140, 1147]

# adjust the format of the income table
# merge first three columns
X = income_table.copy()
X['trends_query_with_cat_id'] = 'US-' + X['state_id'] + '-' + X['id'].astype(str)
X = X.drop(columns=['state_id', 'id', 'name'])

# melt the table
X = X.melt(id_vars=['trends_query_with_cat_id'], var_name='year', value_name='income')
# rename the column year values with lambda
X['year'] = X['year'].map(lambda x: x.split('_')[-1])
# merge year and trends_query_with_cat_id
X['trends_query_with_cat_id'] = X['trends_query_with_cat_id'] + '-' + X['year']
# drop year
X = X.drop(columns=['year'])
# make categories list a dataframe
categories_df = pd.DataFrame(categories, columns=['cat_id'])

# cross join
X['-1'] = 0
categories_df['-1'] = 0
X = X.merge(categories_df, on='-1', how='outer')
X = X.drop(columns=['-1'])

# merge the cat_id column with trends_query_with_cat_id
X['trends_query_with_cat_id'] = X['trends_query_with_cat_id'] + '-' + X['cat_id'].astype(str)
# drop the cat_id column
X = X.drop(columns=['cat_id'])

income_table = X.copy()
income_table
```

```python
X = google_trends_data.copy()

# convert all the row/column indexing to rows

X = X.stack().reset_index().rename(
    columns={
        'level_0': 'cat_idx',
        'level_1': 'trends_query',
    }
)
# map the category idx to category id
X['cat_id'] = X['cat_idx'].map(lambda x: categories[x])
# drop the cat_idx column
X = X.drop(columns=['cat_idx'])
# merge two columns
X['trends_query'] = X['trends_query'] + '-' + X['cat_id'].astype(str)
# drop the cat_id column
X = X.drop(columns=['cat_id'])
# rename the trends_query column 
X = X.rename(columns={'trends_query': 'trends_query_with_cat_id'})
# rename the 0 column
X = X.rename(columns={0: 'trends_time_series'})

google_trends_data = X.copy()
google_trends_data
```

```python
# We have a list of values in the column "trends_time_series", and we want to convert this list into a column with name "time" and "trend_value"

series = google_trends_data['trends_time_series'].apply(pd.Series)
series_stacked = series.stack().reset_index()
series_stacked = series_stacked.rename(
    columns={
        0: 'trends_value',
        'level_0': 'idx',
        'level_1': 'time',
    }
)

series_stacked

# make the index column
google_trends_data['idx'] = google_trends_data.index

# merge on idx
google_trends_data = google_trends_data.merge(series_stacked, on='idx')

# drop the idx and trends_time_series columns
google_trends_data = google_trends_data.drop(columns=['idx', 'trends_time_series'])

google_trends_data
```

```python
# remove the rows that do not exist in the google trends data
income_table = income_table[income_table['trends_query_with_cat_id'].isin(google_trends_data['trends_query_with_cat_id'])]

# sort by trends_query_with_cat_id
income_table = income_table.sort_values(by=['trends_query_with_cat_id'])

# reindex
income_table = income_table.reset_index(drop=True)

# remove the rows that do not exist in the income table
google_trends_data = google_trends_data[google_trends_data['trends_query_with_cat_id'].isin(income_table['trends_query_with_cat_id'])]

# sort by trends_query_with_cat_id and time
google_trends_data = google_trends_data.sort_values(by=['trends_query_with_cat_id', 'time'])

# reindex
google_trends_data = google_trends_data.reset_index(drop=True)
```

```python
# extract all the features
all_extracted_features = tsfresh.extract_features(google_trends_data, column_id='trends_query_with_cat_id', column_sort='time')
all_extracted_features = all_extracted_features.reset_index()
all_extracted_features = all_extracted_features.rename(columns={'index': 'trends_query_with_cat_id'})
all_extracted_features = all_extracted_features.dropna(axis=1)
all_extracted_features
```

```python
# map the trends_query_with_cat_id column to integer hash value
Y = pd.Series(income_table['income'])
Y
```

```python
X = google_trends_data.copy()
X['trends_query_with_cat_id'] = X['trends_query_with_cat_id'].map(lambda x: income_table[income_table['trends_query_with_cat_id'] == x].index[0])
X
```

```python
# make income column to be pandas series
relevant_and_significant_features = tsfresh.extract_relevant_features(X, Y,
                                                     column_id='trends_query_with_cat_id', column_sort='time')
relevant_and_significant_features
```

```python
# save everything
# make sure data/ folder exists
os.makedirs('data', exist_ok=True)

final_table_all = all_extracted_features.merge(income_table, on='trends_query_with_cat_id', how='outer')
final_table_relevant = pd.concat([income_table, relevant_and_significant_features], axis=1)

# save as csv
final_table_all.to_csv('data/final_table_all.csv', index=False)
final_table_relevant.to_csv('data/final_table_relevant.csv', index=False)
```
